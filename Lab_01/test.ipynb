```python
def convert_month(month: str) -> int: 
    if (month == "jan"):
        month = 1
    elif (month == "feb"):
        month = 2
    elif (month == "mar"):
        month = 3
    elif (month == "apr"):
        month = 4
    elif (month == "may"):
        month = 5
    elif (month == "jun"):
        month = 6
    elif (month == "jul"):
        month = 7
    elif (month == "aug"):
        month = 8
    elif (month == "sep"):
        month = 9
    elif (month == "oct"): 
        month = 10
    elif (month == "nov"):
        month = 11
    elif (month == "dec"):
        month = 12 

    return month

df["month"] = df["month"].apply(convert_month)
```

```python
def convert_day(day: str) -> int: 
    if (day == "sun"):
        day = 1
    elif (day == "mon"):
        day = 2
    elif (day == "tue"):
        day = 3
    elif (day == "wed"):
        day = 4
    elif (day == "thu"):
        day = 5
    elif (day == "fri"):
        day = 6
    elif (day == "sat"):
        day = 7
    
    return day
        
df["day"] = df["day"].apply(convert_day)
```

```python
class LinearRegression: 
    def rmse(self, y_hat: np.ndarray, y: np.ndarray) -> float: 
        e = y_hat - y 
        return np.sqrt(np.mean(e**2))
    
    def fit(self, X: np.ndarray, y: np.ndarray) -> None:
        cov = X.T @ X
        inv_cov = np.linalg.inv(cov)
        self.theta_: np.ndarray = inv_cov @ (X.T @ y)
    
    def predict(self, X: np.ndarray) -> np.ndarray:         
        return X @ self.theta_
            
```

```python
lr = LinearRegression()
```
